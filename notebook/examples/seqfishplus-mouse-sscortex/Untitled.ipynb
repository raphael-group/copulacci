```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource
import numpy as np

# Enable Bokeh in Jupyter Lab
output_notebook()

# Sample data
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
desc = [f'Point {i+1}' for i in range(100)]  # Description for each point

# Prepare data source for Bokeh plot
source = ColumnDataSource(data=dict(
    x=x,
    y=y,
    desc=desc,
))

# Create a new plot with a title and axis labels
p = figure(title="Scatter plot with hover tool", x_axis_label='X', y_axis_label='Y', tools="")

# Add a scatter renderer with a size, color, and alpha
p.circle('x', 'y', size=10, source=source, color="navy", alpha=0.5)

# Add HoverTool to the plot
hover = HoverTool()
hover.tooltips = [
    ("Index", "$index"),
    ("(X,Y)", "($x, $y)"),
    ("Description", "@desc"),
]
p.add_tools(hover)

# Show the result
show(p)

```

```python
data = {
    'Column1': [10, -20, 30, -40, 50],
    'Column2': [50, 20, -30, 40, -10]
}
df = pd.DataFrame(data)

# Rank the absolute values in descending order
df['Rank1'] = df['Column1'].abs().rank(method='dense', ascending=False)
df['Rank2'] = df['Column2'].abs().rank(method='dense', ascending=False)
```

```python
res = pd.read_csv('l6_l5.csv',index_col=0)
```

```python
source = ColumnDataSource(data=dict(
    x=res.copula_coeff,
    y=res.scc,
    desc=res.index,
))
p = figure(title="Scatter plot for Copula vs SCC", x_axis_label='Copula', y_axis_label='SCC', tools="")
p.circle('x', 'y', size=5, source=source, color="navy", alpha=0.5)
hover = HoverTool()
hover.tooltips = [
    ("(copula,scc)", "($x, $y)"),
    ("LR_index", "@desc"),
]
p.add_tools(hover)

# Show the result
show(p)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example start points
X = np.array([0])
Y = np.array([0])

# Example vectors (dx, dy) indicating direction and magnitude of the edges
U = np.array([0.5])  # Change in X
V = np.array([0.5])  # Change in Y

# Create a figure and axis
fig, ax = plt.subplots()

# Use quiver to draw directed edges (vectors)
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, width=0.004,headwidth=10, headlength=10, headaxislength=10)

# Setting the axes limits
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

ax.axhline(0.5)
ax.axvline(0.5)
# Show the plot
plt.show()
```
